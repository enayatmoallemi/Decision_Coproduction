## Load and process case data

```python
##Em
# Set spreadsheet directory
directory = 'N:/Planet-A/Current-Users/Enayat-Moallemi/Human_Policy_Integration/Data/'

dimensions = ['Framing', 'Solving', 'Acting', 'Actor', 'Timing', 'Interaction', 
              'Power', 'Politics', 'Change']

cases_dict = {}


# Read and process spreadsheets and drop unnecessary columns
for d in dimensions:
    cases_dict[d] = pd.read_excel(directory+'Supplementary_Data_3.xlsx', sheet_name=d)
    df = cases_dict[d]
    df[d] = df[df.columns[1:]].apply(
        lambda x: ','.join(x.dropna().astype(str)),
        axis=1)
    cases_dict[d] = cases_dict[d].drop(cases_dict[d].columns[[1, 2, 3, 4]], axis=1)
    
results_df = cases_dict['Framing']
for d in dimensions[1:]:
    results_df = pd.merge(results_df,cases_dict[d][['Authors (Year)',d]],on='Authors (Year)', how='left')

scores_df = results_df.iloc[:, 1:].apply(pd.to_numeric).copy()
cases_df = results_df.loc[:, ['Authors (Year)']].copy()
```

## Cluster cases

```python
n_clusters=4
clusterer = KMeans(n_clusters=n_clusters, random_state=10)
labels_used = clusterer.fit_predict(X)
cases_df['Cluster'] = labels_used
scores_df['Cluster'] = labels_used
results_df['Cluster'] = labels_used
```

```python
# Group the cases based on their clusters
df = scores_df.copy()
df2 = results_df.copy()

clusters_dict = {} #without the authors (year) column
clusters_dict2 = {} #with the authors (year) column

for c in range(n_clusters):
    clusters_dict[c]=df.loc[df['Cluster'] == c]
    clusters_dict[c] = clusters_dict[c].drop(['Cluster'], axis=1)
    clusters_dict2[c]=df2.loc[df2['Cluster'] == c]
    clusters_dict2[c] = clusters_dict2[c].drop(['Cluster'], axis=1)
```

## Visualization of the clusters

```python
from matplotlib import ticker
import matplotlib.patches as patches
from matplotlib.path import Path
from matplotlib.patches import BoxStyle
from matplotlib.lines import Line2D
import matplotlib.gridspec as gridspec



cluster_names = {0:'Collaboration', 1:'Innovation', 2:'Transformation', 3:'Implementation'} 
cluster_colors = {0:'#FF7600', 1:'#37E2D5', 2:'#FF79CD', 3:'#FBCB0A'}

levels = ['Level 1', 'Level 2', 'Level 3', 'Level 4']
x_labels = ['Framing', 'Solving', 'Acting', 'Actor', 'Timing', 'Interaction', 'Power', 'Politics', 'Change']


# Create grids and axes
fig = plt.figure(figsize=(20, 12))
grid1 = plt.GridSpec(2, 2, hspace=0.08, wspace=0.05)
axes = {}

# iterate over clusters
for c in range(n_clusters):
    
    # this setting is  for ploting the archetypes in a certain order.
    '''if c < 2:
        gi = c #row
        gj = 0 #column
    else:
        gi = c-2
        gj = 1'''
    
    if c == 0:
        gi = 1 #row
        gj = 0 #column
    if c == 1:
        gi = 0 #row
        gj = 0 #column
    if c == 2:
        gi = 1 #row
        gj = 1 #column
    if c == 3:
        gi = 0 #row
        gj = 1 #column
        
    

    # List the objectives   
    objs = dimensions
    # Create an array of integers ranging from 0 to the number of objectives                    
    x = [i for i, _ in enumerate(objs)]
    
   
    clusters_dict_copy = clusters_dict[c].copy()
    np_indices = clusters_dict_copy.to_numpy()  
   
    
    # Create axes
    for o in range(len(objs)-1):
        grid2 = gridspec.GridSpecFromSubplotSpec(1, 8, wspace=0.0, subplot_spec=grid1[gi,gj])

        axes[c,o] = fig.add_subplot(grid2[o])
        axes[c,o].set_ylim([0.8, 4.2])
            
        # In each cluster and each objective (dimension), count how many studies of each level exist. 
        # unique1 and counts1 show the unique levels and their frequency (respectively)  
        b1 = np_indices[:,o]
        unique1, counts1 = np.unique(b1, return_counts=True)
        
        
        # In each cluster and each "next" objective (dimension), count how many studies of each level exist. 
        # Calculating this second objective is important for defining the line width,= later.  
        b2 = np_indices[:,o+1]
        unique2, counts2 = np.unique(b2, return_counts=True)
        
        # Iterate through all cases
        for d in range(len(np_indices)):
            
            # To define the linewidth between two objective, find how many studies at each level in each objective exist
            # then set their minimum as their linewidth. This is a solution for the line overlay issues where we don't see how many studies are at each level.
            itemindex1 = np.where(unique1==np_indices[d,o])
            itemindex2 = np.where(unique2==np_indices[d,o+1])
            lw = min(counts1[itemindex1[0][0]], counts2[itemindex2[0][0]])
            
            axes[c,o].plot(objs, np_indices[d, :], scaley=True, color=cluster_colors[c], 
                               alpha=1, linewidth=lw)
        axes[c,o].set_xlim([x[o], x[o+1]])
        axes[c,o].set_xticks([o])
        if c in [0, 2]: 
            axes[c,o].set_xticklabels([x_labels[o]], rotation=30, ha='right', size=12)
        else:
            axes[c,o].set_xticklabels([])
      
        # Iterate over and label y axes and set them according to the  levels
        ticks = [1, 2, 3, 4]
        tick_labels = levels
        if (o==0):
            axes[c,o].yaxis.set_ticks(ticks)
            axes[c,o].margins(y=.1)
            axes[c,o].spines.right.set_visible(False)
            axes[c,o].set_yticklabels([], fontsize=10)
        else:
            axes[c,o].yaxis.set_ticks(ticks)
            axes[c,o].set_yticklabels([], fontsize=12)
            axes[c,o].set_yticks([])
            axes[c,o].spines.right.set_visible(False)
            axes[c,o].spines.left.set_visible(False)
            
        if (o==0) & (c in [0,1]):
            axes[c,o].set_yticklabels(tick_labels, fontsize=10)
        
            
        
    # Create a twin axis on the last axis in ax
    ax2 = plt.twinx(axes[c,7])
    ax2.xaxis.set_major_locator(ticker.FixedLocator([x[-2], x [-1]]))
    ax2.set_ylim([0.8, 4.2])
    ax2.spines.left.set_visible(False)
    ticks = [1, 2, 3, 4]
    tick_labels = []
    ax2.yaxis.set_ticks(ticks)
    ax2.set_yticks([])
    ax2.set_yticklabels(tick_labels)
    #ax2.set_xticks([8])
    #ax2.set_xticklabels([x_labels[8]], rotation=90, ha='right')
    axes[c,o].set_xticks([o, o+1])
    
        
    
    if c in [0, 1]:
        axes[c,0].set_ylabel(u"Decision co-production \n higher maturity \u2192", fontsize=12)
        
    if c in [0, 2]:    
        axes[c,o].set_xticklabels([x_labels[o], x_labels[o+1]], rotation=30, ha='right', size=12)
        
    

plt.savefig('{}/Parallelcoordinate_{}_600dpi.png'.format(r'N:/Planet-A/Current-Users/Enayat-Moallemi/Human_Policy_Integration/Fig/', n_clusters), dpi=600,  bbox_inches='tight', transparent=True) 
#plt.savefig('{}/Pareto_Set_Parallelcoordinate.pdf'.format(r'N:/Planet-A/Current-Users/Enayat-Moallemi/Food_Meta_analysis/Figs/'), dpi=600,  bbox_inches='tight') 
#plt.savefig('{}/Pareto_Set_Parallelcoordinate.png'.format(r'N:/Planet-A/Current-Users/Enayat-Moallemi/Food_Meta_analysis/Figs/'), dpi=600,  bbox_inches='tight') 
```
